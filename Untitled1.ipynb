```python
log=[]
def users():
    while True:
        no=int(input("Enter a number of user you want to create:"))
        for i in range(no):
            username=input(f"Enter a user {i+1} name:").lower()
            password=input(f"Enter a {i+1} Password:")
            log.append([username,password])
        check=input("What you want to check: \n1: Count of users \n2:want to search any user? \n3:want to remove any user? \n4: want to add any user? \nPress any for exit:")
        if check=="1":
            print(f"Total Number of Users are: {len(log)}")
            break
        elif check=="2":
            lst=[]
            for i in range(len(log)):
                lst.append(log[i][0])
            check2=input("Enter a name of user you want to search:").lower()
            count=lst.count(check2)
            if count>0:
                print("User Exist!")
                break
            else:
                print("User Does not Exist!")
                break
        elif check=="3":
            lst=[]
            for i in range(len(log)):
                lst.append(log[i][0])
            check2=input("Enter a name of user you want to remove?").lower()
            var=lst.index(check2)
            print(log.pop(var))
            print(f"updated User details are :{log}")
            
            
        elif check == "4":
            continue
        else:
            print("Thank you for using Admin portal!")
            break
                
        break
            
```

```python
course=[]
def courses():
    while True:
        num=int(input("Enter a number of courses you want to add?:"))
        if num>=6:
            print("You have exceeded the limit of courses:!")
            break
        else:
            all=[]
            for i in range(1,num+1):
                name=input("Enter a name of course you want to add:")
                all.append(name)
                print(all)
            course.append(all)
            print(course)  
            
        rerun=input("Do you want to add courses for further students? (yes or no):").lower()
        if rerun=="yes":
            continue
        else:

            break
                
                
```
